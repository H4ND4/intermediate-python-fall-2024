```python
def drop_empty(df):
    return df.dropna(axis=1)
```

```python
def fill_empty(df, col):
    return df.fillna(df[col].mean())
```

```python
fill_empty(df, 'bonus')
```

```python
def drop_column(df, col):
    return df.drop(columns=[col])
```

```python
drop_column(df, 'bonus')
```

```python
def fix_index(df):
    return df.reset_index(drop=True)
```
