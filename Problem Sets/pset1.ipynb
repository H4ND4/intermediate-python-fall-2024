```python
# Your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
# Load a sample Titanic dataset
titanic_data = sns.load_dataset('titanic')
```

```python
# Let's create the Plot class as described
class Plot:
    def __init__(self, df):
        """
        Initialize the class with the dataframe.
        """
        self.df = df

    def box_plots(self):
        """
        This function generates box plots for all numeric features in the dataframe.
        """
        numeric_cols = self.df.select_dtypes(include='number').columns
        self.df[numeric_cols].plot(kind='box', subplots=True, layout=(2, len(numeric_cols)//2), figsize=(12,8), title='Box Plots of Numeric Features')
        plt.tight_layout()
        plt.show()

    def scatter_plot(self, col1, col2):
        """
        Generates a scatter plot for two specific columns.
        """
        if col1 in self.df.columns and col2 in self.df.columns:
            self.df.plot.scatter(x=col1, y=col2, title=f'Scatter plot between {col1} and {col2}')
            plt.show()
        else:
            print(f"Columns {col1} or {col2} not found in dataframe.")

    def scatter_matrix(self):
        """
        Plots a scatter matrix for numeric columns with histograms on the diagonal.
        """
        numeric_cols = self.df.select_dtypes(include='number').columns
        if len(numeric_cols) > 0:
            scatter_matrix(self.df[numeric_cols], figsize=(10, 10), diagonal='hist')
            plt.show()
        else:
            print("No numeric columns available for scatter matrix.")
```

```python
# Create an instance of the Plot class
plot_instance = Plot(titanic_data)
```

```python
# Box Plots
plot_instance.box_plots()
```

```python
# Scatter Plot between 'age' and 'fare'
plot_instance.scatter_plot('age', 'fare')
```

```python

# Scatter Matrix
plot_instance.scatter_matrix()
```
